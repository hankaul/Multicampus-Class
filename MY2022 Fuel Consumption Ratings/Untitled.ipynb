- Model Year : 차량 연식

- Make  : 차량 제조사

- Model : 차량 모델 유형: 4WD/4X4 = 4륜 구동; AWD = 전 륜구동; FFV = 유연한 연료 차량; SWB = 짧은 휠베이스; LWB = 긴 휠베이스; EWB = 확장

- Vehicle Class : 자동차 차량 클래스


- Engine size(L) : 엔진 크기(리터)

- Cylinders : 실린더 수


- Transmission : 변속기: A = 자동; AM = 자동 수동; AS = 선택 시프트로 자동; AV = 연속 가변; M = 수동; 3 – 10 = 기어 수


- Fuel Type : 연료 종류


- Fuel Consumption (City (L/100 km) : 도시 및 고속도로 연료 소비량 등급은 100km당 리터(L/100km)로 표시됩니다

- Fuel Consumption(Hwy (L/100 km)) : 결합된 등급(55% 도시, 45% 고속도로)은 L/100km

- Fuel Consumption(Comb (L/100 km))

- Fuel Consumption(Comb (mpg)) : 결합된 등급(55% 도시, 45% 고속도로)은 갤런당 마일

- CO2 Emissions(g/km) : CO2 배출량: 도심 및 고속도로 결합 시 배기관의 이산화탄소 배출량(킬로미터당 그램 단위)


- CO2 Rating : CO2 등급: 1(최악)에서 10(최상)까지 등급이 지정된 이산화탄소의 배기관 배출


- Smog Rating : 스모그 등급: 1(최악)에서 10(최상)까지 등급이 매겨진 스모그 형성 오염물질의 배기관 배출

### wilcox검정

```python
stat,p = stats.wilcoxon(df.loc[df['Fuel Type']=='Z', 'CO2_Emissions'][:-12],
                         df.loc[df['Fuel Type']=='X', 'CO2_Emissions'])

print(f't통계량 : {round(stat,2)},\np-value : {round(p,2)}')
```

```python
stat,p = stats.ttest_ind(df.loc[df['Fuel Type']=='Z', 'CO2_Emissions'],
                         df.loc[df['Fuel Type']=='X', 'CO2_Emissions'])

print(f't통계량 : {round(stat,2)},\np-value : {round(p,2)}')
```

## 모델 선택 : DNN

```python
df = pd.get_dummies(df,columns=['Fuel Type'])
```

```python
df_X = df.drop(['Make','Model','Vehicle Class','CO2_Emissions','Transmission'],1)
```

```python
df_y = df[['CO2_Emissions']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_set,X_test,y_train_set,y_test = train_test_split(df_X,df_y,random_state=123, test_size=0.1)
```

```python
X_train,X_valid,y_train,y_valid= train_test_split(X_train_set,y_train_set,random_state=123,test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import ModelCheckpoint,ReduceLROnPlateau
from tensorflow.keras.utils import plot_model
```

```python
input_1 = Input(shape=(12,))
hidden_1 = Dense(512,activation='relu')(input_1)
hidden_2 = Dense(256,activation='relu')(hidden_1)
hidden_3 = Dense(128,activation='relu')(hidden_2)
hidden_4 = Dense(64,activation='relu')(hidden_3)
hidden_5 = Dense(32,activation='relu')(hidden_4)
hidden_6 = Dense(16,activation='relu')(hidden_5)
output_1 = Dense(1,activation ='linear')(hidden_6)
model=Model(inputs=input_1,outputs=output_1)
```

```python
plot_model(model)
```

```python
model.summary()
```

```python
import os
os.getcwd()
```

```python
ceckpoint_path = 'my_checkpoint.hdf5'

CP = ModelCheckpoint(filepath = ceckpoint_path,
                             save_weights_only = True,
                             save_best_only = True,
                             monitor = 'val_loss' #검증세트에 대한 loss 기준으로
                             ,verbose=1)

LR = ReduceLROnPlateau(monitor='val_loss',factor=0.8,patience=3, verbose=1, min_lr=1e-8)

CALLBACK = [CP,LR]
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(optimizer=Adam(lr=1e-3),
                     loss='mse',metrics=['acc'])
```

```python
history= model.fit(x=X_train,y=y_train,validation_data=(X_valid,y_valid)
          ,epochs=1200
          ,callbacks=CALLBACK)
```
